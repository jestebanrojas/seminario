# CARGAR Y PROCESAR INFORMACIÓN DEL CLIMA

```python
#Subir csv del clima previamente organizado de la pagina de la NASA
#En este caso no es posible agregarlo desde el repositorio de Github debido al tamaño del archivo
climatologicos_df=pd.read_csv('/content/Datasets_salida/climatologicos/clima.csv', sep=',')
```

```python
#Concimiento del dataset del clima
climatologicos_df
```

```python
#Cambiar caracteres especiles de la columna Municipio del dataset clima y que de este modo coincidan con las provenientes del dataset de usuarios
climatologicos_df['Municipio'] = climatologicos_df['Municipio'].str.replace('Á', 'A')
climatologicos_df['Municipio'] = climatologicos_df['Municipio'].str.replace('É', 'E')
climatologicos_df['Municipio'] = climatologicos_df['Municipio'].str.replace('Í', 'I')
climatologicos_df['Municipio'] = climatologicos_df['Municipio'].str.replace('Ó', 'O')
climatologicos_df['Municipio'] = climatologicos_df['Municipio'].str.replace('Ñ', 'N')
climatologicos_df['Municipio'] = climatologicos_df['Municipio'].str.replace('Ü', 'U')
climatologicos_df['Municipio'] = climatologicos_df['Municipio'].str.replace('MEDELLIN', 'MEDELLÍN')
```

# CARGAR INFORMACIÓN DE LOS USUARIOS Y PROCESARLA

```python
#Subir csv de los usuarios a la red de energia
usuarios_df=pd.read_csv(path+'usuarios/usuarios.csv', sep=',')
```

```python
#Concimiento del dataset de los usuarios
usuarios_df
```

```python
#Crear la columna total usuarios, que es la suma de los usuarios residenciales y no residenciales
usuarios_df['total_usuarios']=usuarios_df['Total Residencial']+usuarios_df['Total No Residencial']

#Eliminar columnas no necesiarias debido a que ya se encuentra su información sumada en usuarios totales
usuarios_df=usuarios_df.drop(['Estrato 1', 'Estrato 2', 'Estrato 3','Estrato 4','Estrato 5','Estrato 6','Industrial','Comercial','Oficial','Otros'], axis=1)

#Revisar dataset usuarios luego de modificaciones
usuarios_df
```

# Unir Clima y Usuarios y realizar ponderación

Se busca realizar una ponderación para cada una de las variables climatológicas según la cantidad de usuarios de cada municipio, para obtener al final solo un valor para todo el departamento para cada hora.

```python
#Unión del dataset clima y usuarios
climausuarios_df = pd.merge(climatologicos_df, usuarios_df,  how='left', left_on=['Municipio','YEAR','MO'], right_on = ['Municipio','Año','Mes'])
```

```python
print("\n -------------------- ")
print("\n -------------------- ")
print("Tipos de datos del dataframe.")
print("\n -------------------- ")
print("\n -------------------- ")

climausuarios_df.info()
```

```python
#Se crean nuevas columnas multiplicando cada columna del dataset clima por la cantidad total de los usuarios.
#Esto para hacer una ponderación y darle más peso a los municipios con más usuarios
climausuarios_df['ALLSKY_SFC_SW_DWN_pond'],climausuarios_df['ALLSKY_KT_pond'],climausuarios_df['T2M_pond'],climausuarios_df['RH2M_pond'],climausuarios_df['PRECTOTCORR_pond'],climausuarios_df['CLRSKY_SFC_SW_DWN_pond'],climausuarios_df['T2MWET_pond']=climausuarios_df['ALLSKY_SFC_SW_DWN']*climausuarios_df['total_usuarios'],climausuarios_df['ALLSKY_KT']*climausuarios_df['total_usuarios'],climausuarios_df['T2M']*climausuarios_df['total_usuarios'],climausuarios_df['RH2M']*climausuarios_df['total_usuarios'],climausuarios_df['PRECTOTCORR']*climausuarios_df['total_usuarios'],climausuarios_df['CLRSKY_SFC_SW_DWN']*climausuarios_df['total_usuarios'],climausuarios_df['T2MWET']*climausuarios_df['total_usuarios']
climausuarios_df.head(50)
```

```python
#Se agrupa infromación por año, mes, día y hora sumando las filas en cada una de las variables y posteriormente dividir con el total de usuarios y obtener el ponderado
climausuarios_df=climausuarios_df.groupby(['YEAR', 'MO','DY','HR']).agg({'Total Residencial': 'sum',
                                                               'Total No Residencial': 'sum',
                                                               'total_usuarios': 'sum',
                                                               'ALLSKY_SFC_SW_DWN_pond': 'sum',
                                                               'ALLSKY_KT_pond': 'sum',
                                                               'T2M_pond': 'sum',
                                                               'RH2M_pond': 'sum',
                                                               'PRECTOTCORR_pond': 'sum',
                                                               'CLRSKY_SFC_SW_DWN_pond': 'sum',
                                                               'T2MWET_pond': 'sum',
                                                                })
climausuarios_df=climausuarios_df.reset_index()
```

```python
#Se divide las columnas creadas para la ponderación por el total de los usuarios, de esta forma los valores del clima toman valores reales con promedio ponderado
climausuarios_df['ALLSKY_SFC_SW_DWN_pond']=climausuarios_df['ALLSKY_SFC_SW_DWN_pond']/climausuarios_df['total_usuarios']
climausuarios_df['ALLSKY_KT_pond']=climausuarios_df['ALLSKY_KT_pond']/climausuarios_df['total_usuarios']
climausuarios_df['T2M_pond']=climausuarios_df['T2M_pond']/climausuarios_df['total_usuarios']
climausuarios_df['RH2M_pond']=climausuarios_df['RH2M_pond']/climausuarios_df['total_usuarios']
climausuarios_df['PRECTOTCORR_pond']=climausuarios_df['PRECTOTCORR_pond']/climausuarios_df['total_usuarios']
climausuarios_df['CLRSKY_SFC_SW_DWN_pond']=climausuarios_df['CLRSKY_SFC_SW_DWN_pond']/climausuarios_df['total_usuarios']
climausuarios_df['T2MWET_pond']=climausuarios_df['T2MWET_pond']/climausuarios_df['total_usuarios']
```

```python
#Se hace un filtrado de tomar todos los valores diferentes al 2017, ya que nuestro dataset inicia del el 01-01-2018
climausuarios_df = climausuarios_df[climausuarios_df.YEAR!=2017]
climausuarios_df.head(50)
```

```python
print("\n -------------------- ")
print("\n -------------------- ")
print("Tipos de datos del dataframe con variables climatologicas ponderadas.")
print("\n -------------------- ")
print("\n -------------------- ")
climausuarios_df.info()
```

# CARGAR Y PROCESAR INFORMACIÓN DE LA TRM HISTÓRICA

```python
#Cargar Dataset TRM para unirlo al Dataset anterior
TRM_df=pd.read_excel(path+'Macroeconomicos/TRM.xlsx')
TRM_df
```

```python
#Organizar y filtrar columna de tiempo, para que inicie desde el 2018
TRM_df['Fecha (dd/mm/aaaa)'] = pd.to_datetime(TRM_df['Fecha (dd/mm/aaaa)'])
TRM_df = TRM_df[TRM_df['Fecha (dd/mm/aaaa)'] >= '2018-01-01']
TRM_df = TRM_df.reset_index(drop=True)
TRM_df
```

```python
#Separar la columna de fecha en las columnas año, mes y día, de este modo se podrá enlazar al dataset resultante en los pasos anteriores
TRM_df['Fecha (dd/mm/aaaa)'] = TRM_df['Fecha (dd/mm/aaaa)'].astype(str)
TRM_df[['YEAR', 'MO', 'DY']] = TRM_df['Fecha (dd/mm/aaaa)'].str.split('-', expand=True)
TRM_df[['YEAR', 'MO', 'DY']] = TRM_df[['YEAR', 'MO', 'DY']].astype('int64')
TRM_df = TRM_df.drop(['Fecha (dd/mm/aaaa)'], axis=1)
TRM_df
```

# Unir ponderado clima-usuarios con TRM

```python
# Unir el Dataframe de clima-usuarios (organizado previamente) con el dataframe TRM
climausuariosTRM_df = pd.merge(climausuarios_df, TRM_df,
                            how='left', left_on=['YEAR','MO', 'DY'],
                            right_on = ['YEAR','MO', 'DY'])

climausuariosTRM_df
```

# CARGAR Y PROCESAR INFORMACIÓN HISTÓRICA DEL IPP Y DEL IPC

```python
#Cargar Dataset IPC y IPP para unirlo al Dataset anterior
IPC_IPP_df=pd.read_excel(path+'Macroeconomicos/Evolución IPC E IPP base 2008 - Resolución_ Mensual.xlsx')
IPC_IPP_df
```

```python
#Modificar columna de tiempo ya que venía con un formato especial y que pueda ser convertida a formato datetime
meses_personalizados = {
    'ene.': 'Jan',
    'feb.': 'Feb',
    'mar.': 'Mar',
    'abr.': 'Apr',
    'may.': 'May',
    'jun.': 'Jun',
    'jul.': 'Jul',
    'ago.': 'Aug',
    'sep.': 'Sep',
    'oct.': 'Oct',
    'nov.': 'Nov',
    'dic.': 'Dec'
}

# Reemplazar las abreviaturas personalizadas por las estándar
IPC_IPP_df['Fecha'] = IPC_IPP_df['Fecha'].replace(meses_personalizados, regex=True)

# Convierte las fechas al nuevo formato 'YYYY-MM'
IPC_IPP_df['Fecha'] = pd.to_datetime(IPC_IPP_df['Fecha'], format='%b/%Y').dt.strftime('%Y-%m')
```

```python
#Separar la columna de fecha en las columnas año y mes para poder enlazar con el dataframe  climausuariosTRM_df
IPC_IPP_df['Fecha'] = IPC_IPP_df['Fecha'].astype(str)
IPC_IPP_df[['YEAR', 'MO']] = IPC_IPP_df['Fecha'].str.split('-', expand=True)
IPC_IPP_df[['YEAR', 'MO']] = IPC_IPP_df[['YEAR', 'MO']].astype('int64')
IPC_IPP_df = IPC_IPP_df.drop(['Fecha'], axis=1)
IPC_IPP_df
```

# UNIR DATASETS  climausuariosTRM_df CON IPC_IPP_df

```python
# Unir el Dataframe de clima-usuarios-TRM (organizado previamente) con el dataframe IPC e IPP
climausuariosTRMIPCIPP_df = pd.merge(climausuariosTRM_df, IPC_IPP_df,
                            how='left', left_on=['YEAR','MO'],
                            right_on = ['YEAR','MO'])

climausuariosTRMIPCIPP_df
```

```python
print("\n -------------------- ")
print("\n -------------------- ")
print("Información general del dataframe final")
print("\n -------------------- ")
print("\n -------------------- ")

climausuariosTRMIPCIPP_df.info()
```
