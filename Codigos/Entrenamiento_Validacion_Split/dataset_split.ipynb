# ENFOQUE 1

El enfoque 1, consiste en separar un dataset de entrenamiento con los datos entre el 2018-01-01 y el 2022-12-31 y los datos de validación serán los comprendidos entre el 2023-01-01 y el 2023-09-30

```python
#Construir una columna tipo datetime apartir de las columnas AÑO, MES, DIA, HORA
dataset_completo['FECHA'] = pd.to_datetime(dict(year=dataset_completo['AÑO'], month=dataset_completo['MES'], day=dataset_completo['DIA'],hour=dataset_completo['HORA']))

#Separación de los datos de entrenamiento
df_entrenamiento = dataset_completo[(dataset_completo['AÑO'] >= 2018) & (dataset_completo['AÑO'] <= 2022)]

# #Separación de los datos de validación: Filtrar datos para el año 2023
df_validacion = dataset_completo[dataset_completo['AÑO'] == 2023]
```

# ENFOQUE 2


```python
#crear una lista con el nombre de todas las columnas del dataset para la construcción de un dataset X con las variables predictoras
Cols_X = list(dataset_completo.columns)
#Se remueve de la lista el nombre DEMANDA, ya que corresponde a la variable a predecir, es decir y
Cols_X.remove('DEMANDA')
X = dataset_completo[Cols_X]
y = dataset_completo['DEMANDA']


```

```python
X
```

```python
y
```

```python
# Se crean los datasets de entrenamiento y prueba para las variables de entrada y salida
X_train, X_test, y_train, y_test = train_test_split(
                                        X,
                                        y.values.reshape(-1,1),
                                        train_size   = 0.7,
                                        random_state = 1234,
                                        shuffle      = True
                                    )
```

```python
#Se construye nuevamente el dataset con todas las columnas
df_entrenamiento=X_train
df_entrenamiento['DEMANDA']=y_train
```

```python
#Se construye nuevamente el dataset con todas las columnas
df_validacion=X_test
df_validacion['DEMANDA']=y_test
```

```python
#Verificar si existe directorio en el entorno
import os
path = "Datasets_salida/Entrenamiento_Validacion_Split/enfoque2/"
# verificación de si existe o no el directorio
isExist = os.path.exists(path)
if not isExist:

   # Ccrea el directorio cuando no existe
   os.makedirs(path)
   print("\n -------------------- \n")
   print("\n -------------------- \n")
   print("se creó el directorio: ", path)

#Descargar en la carpeta de drive los Dataset de entrenamiento y validación en csv con la información final


df_entrenamiento.to_csv('/content/drive/MyDrive/Trabajo/Datasets_salida/Entrenamiento_Validacion_Split/enfoque2/entrenamiento_df.csv', index=False)
df_validacion.to_csv('/content/drive/MyDrive/Trabajo/Datasets_salida/Entrenamiento_Validacion_Split/enfoque2/validacion_df.csv', index=False)


df_entrenamiento.to_csv('/content/Datasets_salida/Entrenamiento_Validacion_Split/enfoque2/entrenamiento_df.csv', index=False)
df_validacion.to_csv('/content/Datasets_salida/Entrenamiento_Validacion_Split/enfoque2/validacion_df.csv', index=False)

print("\n -------------------- \n")
print("\n -------------------- \n")
print("se creó el archivo: entrenamiento_df.csv en enfoque 2")
print("se creó el archivo: validacion_df.csv en enfoque 2")
print("\n -------------------- \n")
print("\n -------------------- \n")
```
