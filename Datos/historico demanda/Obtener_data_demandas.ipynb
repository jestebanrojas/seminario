```python
# importar las librerías requeridas
import pandas as pd
import os
import glob

#Importar librería de google drive para la carga de los archivos de datos requeridos.
from google.colab import drive
drive.mount('/content/drive')

# Se crea un path específico para facilitar la lectura de los archivos posteriormente desde google drive
#path = "/content/drive/MyDrive/Trabajo/Datos/historico demanda/historico"

#Se clona el repositorio de Github donde se encuentran los datasets fuente
! git clone https://github.com/jestebanrojas/seminario.git
# Se crea un path específico para facilitar la lectura de los archivos posteriormente desde github
path = "/content/seminario/Datos/historico demanda/historico"
```

```python
#Se almacena las rutas de todos los archivos en el path que coincidan con la extensión xlsx
csv_files = glob.glob(os.path.join(path, "*.xlsx"))

i=0

#Crear un dataframe vacío, el cual se usará para ir anexando la consulta de cada archivo en un solo dataframe
df_total = pd.DataFrame()

#iterar por cada uno de los archivos con la extensión requerida y anexar al dataframe final
for f in csv_files:
    i+=1
    # se lleva a un dataframe la información del archivo
    df = pd.read_excel(f, 'real')

    # imprimir el # del archivo, el path y el nombre del archivo
    print('archivo # ',i)
    print('Location:', f)



    # se anexa el dataframe de la iteración al dataframe que acumula los datos
    df_total=pd.concat([df_total,df],sort=False)

# mostrar el dataset resultante
print('Content:')
display(df_total)
print()
```

```python
df_total.info()
```

```python
#Homogenizar el título de algunas columnas para hacerlo compatible con la información que se descargarpa de los archivos con extensión xls
df_total=df_total.rename({'Varialble' : 'VARIABLE'}, axis=1)
df_total=df_total.rename({'TIPO_DIA' : 'TIPO DIA'}, axis=1)
df_total=df_total.rename({'Total' : 'TOTAL'}, axis=1)
```

```python
#Se almacena las rutas de todos los archivos en el path que coincidan con la extensión xls
csv_files = glob.glob(os.path.join(path, "*.xls"))

i=0

#Crear un dataframe vacío, el cual se usará para ir anexando la consulta de cada archivo en un solo dataframe
df_total2 = pd.DataFrame()

#iterar por cada uno de los archivos con la extensión requerida y anexar al dataframe final
for f in csv_files:
    i+=1
    # se lleva a un dataframe la información del archivo
    df = pd.read_excel(f, 'real')

    # imprimir el # del archivo, el path y el nombre del archivo
    print('archivo # ',i)
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])




    # se anexa el dataframe de la iteración al dataframe que acumula los datos
    df_total2=pd.concat([df_total2,df],sort=False)

# mostrar el dataset resultante
print('Content:')
display(df_total2)
print()


```

```python
#Convertir a formato fecha los datos de la columna "FECHA" pues se requerirá para enlazar con otros dataset en los diferentes notebook que lo requieran
df_total2['FECHA']= pd.to_datetime(df_total2['FECHA'],format = '%d/%m/%Y')
```

```python
#Se unen los dataframe provenientes de archivos xlsx y xls
df_total3=pd.concat([df_total,df_total2],sort=False)
```

```python
#ordenar el dataframe por fecha
df_total3=df_total3.sort_values(by=['FECHA'])
print(df_total3.to_string())
```
