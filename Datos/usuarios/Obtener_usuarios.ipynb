```python
# importar las librerías requeridas
import pandas as pd
import os
import glob

#Importar librería de google drive para la carga de los archivos de datos requeridos.
from google.colab import drive
drive.mount('/content/drive')

# Se crea un path específico para facilitar la lectura de los archivos posteriormente desde google drive
#path = "/content/drive/MyDrive/Trabajo/Datos/usuarios/Mensuales"



#Se clona el repositorio de Github donde se encuentran los datasets fuente
! git clone https://github.com/jestebanrojas/seminario.git
# Se crea un path específico para facilitar la lectura de los archivos posteriormente desde github
path = "/content/seminario/Datos/usuarios/Mensuales"
```

```python
path
```

```python
#from xlrd.formula import unicode

#Se almacena las rutas de todos los archivos en el path que coincidan con la extensión csv
csv_files = glob.glob(os.path.join(path, "*.csv"))

i=0

#Crear un dataframe vacío, el cual se usará para ir anexando la consulta de cada archivo en un solo dataframe
df_total = pd.DataFrame()

#iterar por cada uno de los archivos con la extensión requerida y anexar al dataframe final
for f in csv_files:

    i+=1
    # se lleva a un dataframe la información del archivo
    df = pd.read_csv(f, encoding='latin-1')

    """Debido a la estructura de los archivos csv descargados del SUI se debe
    identificar en cada archivo el año y el mes al que corresponden, estos se
    encuentran como filas al final de cada archivo en la columna municipio,
    por lo que se debe almacenar como una nueva columna del dataframe"""
    df['Mes'] = df.loc[df['Departamento']=='Período'].Municipio.item()
    df['Año'] = df.loc[df['Departamento']=='Año'].Municipio.item()

    # imprimir el # del archivo, el path y el nombre del archivo
    print('archivo # ',i)
    print('Location:', f)
    print('File Name:', f.split("\\")[-1])

    # se anexa el dataframe de la iteración al dataframe que acumula los datos
    df_total=pd.concat([df_total,df],sort=False)



# mostrar el dataset resultante
print('Content:')
display(df_total)
print()
```

```python
df_total.info()
```

```python
#Se filtran los registros no deseados, que corresponde a los header y finales de cada csv
df_total = df_total[df_total.Departamento=='ANTIOQUIA']
```

```python
#Se reemplazan los campos con el texto'ND' por cero para realizar las sumas totalizadoras de usuarios
df_total = df_total.replace('ND', 0)
```

```python
#Conversión de los datos de usuarios a tipo int para poder sumar
df_total = df_total.astype({'Otros': 'int',
                            'Estrato 1': 'int',
                            'Estrato 2': 'int',
                            'Estrato 3': 'int',
                            'Estrato 4': 'int',
                            'Estrato 5': 'int',
                            'Estrato 6': 'int',
                            'Total Residencial': 'int',
                            'Industrial': 'int',
                            'Comercial' : 'int',
                            'Oficial': 'int',
                            'Total No Residencial': 'int'})
```

```python
#Agrupación para totalizar la cantidad de usuarios por municipio (sumar los usuarios de cada empresa comercializadora)
df_total1 = df_total.groupby(['Municipio', 'Año', 'Mes']).agg({'Departamento': 'first',

                                                               'Estrato 1': 'sum',
                                                               'Estrato 2': 'sum',
                                                               'Estrato 3': 'sum',
                                                               'Estrato 4': 'sum',
                                                               'Estrato 5': 'sum',
                                                               'Estrato 6': 'sum',
                                                               'Total Residencial': 'sum',
                                                               'Industrial': 'sum',
                                                               'Comercial' : 'sum',
                                                               'Oficial': 'sum',
                                                               'Otros':'sum',
                                                               'Total No Residencial': 'sum'})
df_total1
```

```python
df_total1=df_total1.reset_index()
```
